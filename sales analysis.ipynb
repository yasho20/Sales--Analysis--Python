```python
df = pd.read_csv("Sales data.csv")
sales_by_region = df.groupby("Region")["TotalSales"].sum()
print(sales_by_region)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
sales_by_product = df.groupby("Product")["TotalSales"].sum()
sales_by_product.plot(kind="bar",color="skyblue")
plt.title("Total sales by product")
plt.xlabel("Product")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df['Date']= pd.to_datetime(df['Date'])
sales_trend= df.groupby("Date")["TotalSales"].sum()
sales_trend.plot(kind="line",marker = "o",color="green")
plt.title("Sales trend over time")
plt.xlabel("Date")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
sales_by_region = df.groupby("Region")["TotalSales"].sum()
sales_by_region.plot(kind="barh",color="hotpink")
plt.title("Total sales by region")
plt.xlabel("Region")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
profit_by_product= df.groupby("Product")["Profit"].sum()
profit_by_product.plot(kind="pie",startangle = 180)
plt.title("Profit share by product")
plt.ylabel("")
plt.show()
```
